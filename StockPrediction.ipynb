```python
# 0 Preprocessing + EDA + Features selection
Df=pd.read_csv('bnfdata.csv')
Df


Df['Average_Price'] = (Df['high'] + Df['low']) / 2
Df['PriceChange']=Df['open']-Df['close']
Df['Next_Close'] = Df['close'].shift(-1)
Df['NewClose']=Df['close']-Df['open'].shift(-1)
```

```python
# 0.3 FS 
Df.drop(['symbol','volume'], axis=1, inplace=True)
# df.drop('column_name', axis=1, inplace=True
```

```python
# 0.3 Fs
Df = Df.iloc[:, 1:]
Df
```

```python
# 1 Extract input and output col
# 2 Scale the value
```

```python
# 3 Train test splite 

from sklearn.model_selection import train_test_split
X=Df[['open','high','low','close','Average_Price','PriceChange','Next_Close']]
# Target
Y=Df['Next_Close']

X_train, X_test, y_train, y_test=train_test_split(X,Y,test_size=0.1)
```

```python
# Scaling the values
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
# training the model
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
```

```python
# 5 Evalute the model / Model Selection      _>>>>> Check the accurcy with test data 

LR.predict(X_train)
```
